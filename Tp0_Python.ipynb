```python
import matplotlib.pyplot as plt
import numpy as np

# Señal cuadrada (escalón)
senial_cuad = [] #inicializo un array vacío
for a in range(5):
    senial_cuad.append(0)
for a in range(5):
    senial_cuad.append(1)
print(senial_cuad)
plt.title("Escalón")
plt.plot(senial_cuad)
plt.show()

# Señal delta de dirac
delta_dirac = []
for a in range(100):
    delta_dirac.append(0)
dd= int(len(delta_dirac)/2) #obtener la posición del medio del array
delta_dirac[dd] = 1
print("el largo del vector delta_dirac es: {}".format(len(delta_dirac)))
plt.title("Dirac")
plt.plot(delta_dirac)
plt.show()

# Señal triangular
senial_tria = []
for a in range(5):
    senial_tria.append(a)
for a in range(5,0,-1):
    senial_tria.append(a)
plt.title("Triangular")
plt.plot(senial_tria)
plt.show()


#Se puede definir funciones y parametrizar las señales que voy a generar
#Ejemplo con delta de dirac
def generar_delta_dirac(cantidad,pos):
    delta_dirac = []
    for a in range(cantidad):
        delta_dirac.append(0)
    delta_dirac[pos] = 1
    return delta_dirac

print(generar_delta_dirac(10,5))
```

```python
# 1. Definir funciones para generar señal cuadrada y triangular en python 'puro' (sin librerías).
def cuadrada_sin_lib (ciclos):
  definicion = 100
  y = [ int((val % definicion)/definicion + 0.5) for val in range(0,definicion * ciclos,1)]
  x = list(range(definicion * ciclos))
  return x, y

#plot
print("x,y: {}".format(cuadrada_sin_lib(10)))
x, y = cuadrada_sin_lib(10)
plt.plot(x,y)
plt.show()

#Triangular
def triangular_sin_lib (ciclos):
  definicion = 100
  y = [ ((val % definicion),-val % definicion)[(val % definicion)>definicion/2] - definicion/2 for val in range(0,definicion * ciclos,1)]
  x = list(range(definicion * ciclos))
  return x, y

#plot
print("x,y: {}".format(triangular_sin_lib(10)))
x, y = triangular_sin_lib(10)
plt.plot(x,y)
plt.show()
```

```python
# 2. Generar, con *numpy* y/o *scipy.signal*, las señales: cuadrada, dirac, triangular, y senoidal.
import numpy as np
import scipy.signal as sig

# Pista: Esto es como lo que hicimos en el punto anterior, pero usando el método
# np.append en lugar del .append propio de Python.

# Señal Cuadrada con numpy
def cuadrada_lib (ciclos):
  definicion = 100
  y = np.array(range(0,definicion * ciclos,1))
  y = ( (y % definicion) / definicion + 0.5).astype(int)
  x = list(range(definicion * ciclos))
  return x, y

#plot
#print("x,y: {}".format(cuadrada_lib(10)))
x, y = cuadrada_lib(10)
plt.plot(x,y)
plt.show()

# Señal delta de dirac con numpy
def delta_dirac_lib (cantidad,pos):
    delta_dirac = np.zeros(cantidad)
    delta_dirac[pos] = 1
    x = list(range(cantidad))
    return x, delta_dirac

#plot
#print("x,y: {}".format(delta_dirac_lib(10,5)))
x, y = delta_dirac_lib(1000,500)
plt.plot(x,y)
plt.show()

# Señal triangular con numpy
def triangular_lib (ciclos):
  definicion = 100
  y = np.array(range(0,definicion * ciclos,1))
  condicion = y % definicion > definicion/2
  y = y - definicion/2
  y = np.where(condicion, y % definicion, -y % definicion)
  x = list(range(definicion * ciclos))
  return x, y

#plot
#print("x,y: {}".format(triangular_lib(10)))
x, y = triangular_lib(10)
plt.plot(x,y)
plt.show()

# Señal senoidal con numpy
def sine_np (ciclos):
  definicion = 100
  y = np.array(range(0,definicion * ciclos,1))
  y = y * np.pi / definicion
  y = np.sin(y)
  x = list(range(definicion * ciclos))
  return x, y

#plot
#print("x,y: {}".format(sine_np(10)))
x, y = sine_np(10)
plt.plot(x,y)
plt.show()
```

```python
# Ejemplo:  Señal cuadrada con scipy
import scipy.signal as sig
import matplotlib.pyplot as plt

#Parámetros de la señal
frec_muestreo = 5000 #en Hz
frec_senial = 2 # en Hz
# Base de tiempo de 2 formas (es indistinto cuál usar)
# t = np.linspace(0, 1, frec_muestreo, endpoint=False)
t = np.arange(0,1,1/frec_muestreo)

print("t: {}".format(t))
senial_cuadrada_scipy = sig.square(2 * np.pi * frec_senial * t)
print("Señal cuadrada scipy: {}".format(senial_cuadrada_scipy))

plt.plot(t, senial_cuadrada_scipy)
plt.ylim(-2, 2)
```

```python
# 3. Hacer lo mismo que en el punto anterior usando una base de tiempo. Graficar todas las señales.
ciclos = 4

# Señal triangular con numpy/scipy? (pista: sig.sawtooth)
x = np.arange(0,ciclos,0.01)
y = sig.sawtooth(2 * np.pi * x, width=0.5)
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Señal Triangular")


# Señal senoidal con numpy/scipy? (pista: np.sin)
x = np.arange(0,ciclos,0.01)
y = np.sin(2 * np.pi * x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Señal Senoidal")


# Delta de dirac con numpy/scipy? (pista: sig.unit_impulse)
x = np.arange(0,ciclos,0.01)
y = sig.unit_impulse(len(x), int(len(x)/2))
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("Señal Delta de Dirac")


#cuadrada
x = np.arange(0,ciclos,0.01)
y = sig.square(2 * np.pi * x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("Señal Cuadrada")


plt.tight_layout()
plt.show()
```

```python
# Sintaxis para el manejo de arrays:
t = np.arange(0,20,1)
print(t)

print(t[-1:]) #obtengo el último valor del array
print(t[0]) #obtengo el primer valor del array
print(t[:]) #obtengo todo el array
print(t[5:7]) #obtengo desde el elemento 5 (inclusive) hasta el 7 (exclusive)
print(t[-5:]) #obtengo los últimos 5 elementos del array
print(t[:3]) #obtengo los primeros 3 elementos del array
print(t[::2]) #elementos en posiciones pares
```

```python
#Desordeno el array t
np.random.shuffle(t)
print(t)

# 4. Calcular máximos y mínimos con subindices en python puro.
min = { "val": t[0], "index": 0}
max = { "val": t[0], "index": 0}
for index, val in enumerate(t):
  if val < min["val"]:
    min["val"] = val
    min["index"] = index
  if val > max["val"]:
    max["val"] = val
    max["index"] = index

print("min: {}".format(min))
print("max: {}".format(max))

# 5. Idem anterior con numpy.
min = { "val": np.min(t), "index": np.argmin(t)}
max = { "val": np.max(t), "index": np.argmax(t)}
print("min: {}".format(min))
print("max: {}".format(max))
```
